```python
## generate fake test dataset 
# index represents genes
# _log FC = from each NASA dataset for each gene
# _per = % match (NCBI)

df = pd.DataFrame(np.random.rand(50, 6) , columns=['PA_logFC','BS_logFC','EC_logFC',
                                                    'PA_per','BS_per','EC_per'])

print(df)
```

```python
df['Total %'] = df['PA_per'] + df['BS_per'] + df['EC_per']

print(df)
```

```python
df['PA_per_N'] = df['PA_per'] / df['Total %']
df['BS_per_N'] = df['BS_per'] / df['Total %']
df['EC_per_N'] = df['EC_per'] / df['Total %']

print(df[:5])
```

```python
df['PA_FCN'] = df['PA_logFC'] * df['PA_per_N']
df['BS_FCN'] = df['BS_logFC'] * df['BS_per_N']
df['EC_FCN'] = df['EC_logFC'] * df['EC_per_N']

print(df[:5])
```

```python
df = df[['PA_FCN','BS_FCN','EC_FCN']]

df['Total_FCN'] = df['PA_FCN'] + df['BS_FCN'] + df['EC_FCN']
#df['Avg_FCN'] = (df['PA_FCN'] + df['BS_FCN'] + df['EC_FCN']) / 3

print(df[:5])
```

```python
df2 = df.copy()

gene = np.arange(0, 50, 1)
df2['Gene'] = pd.DataFrame(gene)

df2.set_index('Gene')
```
